```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("mnist_train.csv")
```

```python
data.head()
```

```python
y = data.pop("label")
x = data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
RF = RandomForestClassifier(n_estimators = 20)
RF.fit(x_train, y_train)
```

```python
print("Random Forest Classifier score : {}".format(RF.score(x_test,y_test)))
```

```python
bagging = BaggingClassifier(DecisionTreeClassifier(), max_samples = 0.5, 
                            n_estimators=20, max_features = 1.0)
bagging.fit(x_train,y_train)
```

```python
print("Bagging ensemble score : {}".format(bagging.score(x_test,y_test)))
```

```python
adaboost = AdaBoostClassifier(DecisionTreeClassifier(criterion = "gini"),
                              n_estimators = 10, learning_rate = 1)
adaboost.fit(x_train,y_train)
```

```python
print("Adaboost score : {}".format(adaboost.score(x_test,y_test)))
```

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
svc = SVC(kernel = 'poly')
vote_classifier = VotingClassifier(estimators = [("lr",lr), ("dt",dt), ("svc",svc)], voting = "hard")
```

```python
vote_classifier.fit(x_train.iloc[1:4000], y_train.iloc[1:4000])
```

```python
print("Vote Classifier score : {}".format(vote_classifier.score(x_test,y_test)))
```
